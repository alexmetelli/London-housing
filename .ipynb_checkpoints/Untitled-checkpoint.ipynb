```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('/home/alex/Documents/London_crime/MPS Ward Level Crime (most recent 24 months).csv')
print('Data shape:{}'.format(data.shape))
data.head()
```

```python
df = data.drop(['WardCode', 'MajorText'], axis=1)
df['Tot'] = df.sum(axis=1)
df.describe()
```

```python
cols = range(3,27)
df.drop(df.columns[cols], axis=1, inplace=True)
df.head()
```

```python
borought = df.pivot_table(index='MinorText',
                   columns='LookUp_BoroughName',
                   values='Tot')
print(borought.shape)
borought.head(56)
```

```python
df['WardName']
```

```python
plt.figure(figsize=(60, 15))
sns.violinplot(x='MinorText', y='Tot', data=df)
```

```python
df_obj = df.select_dtypes(include=['object']).copy()
df_obj.head()
```

```python
df_obj[''].value_counts()
```

```python
pd.get_dummies(df_obj, columns=['MajorText', 'MinorText', 'WardName', 'WardCode', 'LookUp_BoroughName']).head()
```

```python
lb_make = LabelEncoder()
df_obj['MajorText_code'] = lb_make.fit_transform(df_obj['MajorText'])
df_obj['Ward_code'] = lb_make.fit_transform(df_obj['WardName'])
df_obj['borough_code'] = lb_make.fit_transform(df_obj['LookUp_BoroughName'])
```
